```python
import pandas
import numpy as np
import datetime
def score_elo(hp1,hp2,hp3,hp4,csv): # define a new function that takes in hyperparameter values and returns the percent of games correctly predicted
    data = pandas.read_csv(csv)
    ID = {}
    outcomes = []
    winner = {}
    loser = {}
    for idx, row in data.iterrows():
        pd = data.loc[idx,'pd']
        if pd > 0:
            winner[0] = data.loc[idx,'hp1'];
            winner[1] = data.loc[idx,'hp2'];
            winner[2] = data.loc[idx,'hp3'];
            winner[3] = data.loc[idx,'hp4'];
            winner[4] = data.loc[idx,'hp5'];
            loser[0] = data.loc[idx,'ap1'];
            loser[1] = data.loc[idx,'ap2'];
            loser[2] = data.loc[idx,'ap3'];
            loser[3] = data.loc[idx,'ap4'];
            loser[4] = data.loc[idx,'ap5'];
            hm_i = 1
        else:
            winner[0] = data.loc[idx,'ap1'];
            winner[1] = data.loc[idx,'ap2'];
            winner[2] = data.loc[idx,'ap3'];
            winner[3] = data.loc[idx,'ap4'];
            winner[4] = data.loc[idx,'ap5'];
            loser[0] = data.loc[idx,'hp1'];
            loser[1] = data.loc[idx,'hp2'];
            loser[2] = data.loc[idx,'hp3'];
            loser[3] = data.loc[idx,'hp4'];
            loser[4] = data.loc[idx,'hp5'];
            hm_i = 0
        game_date = datetime.datetime.strptime(data.loc[idx]["Date"],"%a, %b %d, %Y")
        for x in winner:
        if winner[x] in ID.keys():
            w_elo[x] = ID[winner][x][0]
            if (ID[winner][x][1] - game_date).days == 1: w_date[x] = 1 
            else: w_date[x] = 0
            w_games[x] = ID[winner][x][2]
        else: 
            w_elo[x] = 1500
            w_date[x] = 0
            w_games[x] = 0
        if loser[x] in ID.keys():
            l_elo[x] = ID[loser][x][0]
            if (ID[loser][x][1] - game_date).days == 1: l_date[x] = 1
            else: l_date[x] = 0
            l_games[x] = ID[loser][x][2]
        else: 
            l_elo[x] = 1500
            l_date[x] = 0
            l_games[x] = 0
        if hm_i == 1: w_elo_g[x] = w_elo[x]+hp3 + hp4*w_date[x]; l_elo_g[x] = l_elo[x] + hp4*l_date[x]
        else: l_elo_g[x] = l_elo[x] +hp3 + hp4*l_date[x]; w_elo_g[x] = w_elo[x] + hp4*l_date[x];
        if w_elo_g[x]>l_elo_g[x]: outcomes.append(1)
        if l_elo_g[x]>w_elo_g[x]: outcomes.append(0)
        w_new_elo[x] = (w_elo[x]+np.log(pd+1)*(2.2/((w_elo[x]-l_elo[x])*.001+2.2))*hp1*(1-(1/(1+10**((l_elo_g[x]-w_elo_g[x])/400)))))
        l_new_elo[x] = (l_elo[x]+np.log(pd+1)*(2.2/((w_elo[x]-l_elo[x])*.001+2.2))*hp1*(0-(1/(1+10**((w_elo_g[x]-l_elo_g[x])/400)))))
        ID[winner][x]=(w_new_elo[x],game_date[x],w_games[x] + 1);
        ID[loser][x]=(l_new_elo[x],game_date[x],l_games[x] + 1)
    print (sum(outcomes)/float(len(outcomes)))
    print(ID)
score_elo(3.9809149872695824, 693.0107331703836, 4.991110715197425, -0.9999998678714583, 'games2015.csv') # run the function on the 2015 data with the given hyperparameters
```
