```python
import pandas
import numpy as np
import datetime
import math
import operator
def score_elo(hp1,hp2,csv): # define a new function that takes in hyperparameter values and returns the percent of games correctly predicted
    data = pandas.read_csv(csv)
    ID = {}
    outcomes = []
    winner = {}
    loser = {}
    for idx, row in data.iterrows():
        pd = data.loc[idx,'pd']
        poss = data.loc[idx,'Possessions']
        t =0 
        if pd > 0:
            winner[0] = data.loc[idx,'hp1'];
            winner[1] = data.loc[idx,'hp2'];
            winner[2] = data.loc[idx,'hp3'];
            winner[3] = data.loc[idx,'hp4'];
            winner[4] = data.loc[idx,'hp5'];
            loser[0] = data.loc[idx,'ap1'];
            loser[1] = data.loc[idx,'ap2'];
            loser[2] = data.loc[idx,'ap3'];
            loser[3] = data.loc[idx,'ap4'];
            loser[4] = data.loc[idx,'ap5'];
        else:
            winner[0] = data.loc[idx,'ap1'];
            winner[1] = data.loc[idx,'ap2'];
            winner[2] = data.loc[idx,'ap3'];
            winner[3] = data.loc[idx,'ap4'];
            winner[4] = data.loc[idx,'ap5'];
            loser[0] = data.loc[idx,'hp1'];
            loser[1] = data.loc[idx,'hp2'];
            loser[2] = data.loc[idx,'hp3'];
            loser[3] = data.loc[idx,'hp4'];
            loser[4] = data.loc[idx,'hp5'];
            if pd ==0:
                t = .5
        w_elo=[]
        l_elo=[]
        w_poss=[]
        l_poss=[]
        for x in winner:
            if winner[x] in ID.keys():
                w_elo.append(ID[winner[x]][0])
                w_poss.append(ID[winner[x]][1])
            else: 
                w_elo.append(1500)
                w_poss.append(0)
            if loser[x] in ID.keys():
                l_elo.append(ID[loser[x]][0])
                l_poss.append(ID[loser[x]][1])
            else: 
                l_elo.append(1500)
                l_poss.append(0)
        w_team  = sum(w_elo)
        l_team = sum(l_elo)
        if w_team>l_team: outcomes.append(1)
        if l_team>w_team: outcomes.append(0)
        for x in winner:
            w_new_elo = (w_elo[x]+.2*poss*hp1*(1-t-(1/(1+10**((l_team-w_team)/400)))))
            l_new_elo = (l_elo[x]+.2*poss*hp1*(0+t-(1/(1+10**((w_team-l_team)/400)))))
            ID[winner[x]]=(w_new_elo,(w_poss[x]+poss));
            ID[loser[x]]=(l_new_elo,(l_poss[x]+poss));
    return ID
ID = score_elo(5, 700, 'rapm.csv') # run the function on the 2015 data with the given hyperparameters
print (sorted(ID.items(),key=operator.itemgetter(1),reverse=True))
```
